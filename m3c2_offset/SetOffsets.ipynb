```python
import pandas as pd
lm = pd.read_csv("Depth_180511.txt")
old_z = lm["Z"].copy()

distance_col = "M3C2 interp"
lm["//X"] -= lm["Nx"] * lm[distance_col]
lm["Y"] -= lm["Ny"] * lm[distance_col]
lm["Z"] -= lm["Nz"] * lm[distance_col]

lm = lm[~lm.Y.isnull()]



lm[["//X", "Y", "Z", distance_col, "Nx", "Ny", "Nz"]].to_csv("Depth_offset_180511.txt", index=False)
lm.head()
```
